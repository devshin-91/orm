```python
# nonlocal 사용
def make_counter():
    count = 0

    def counter():
        nonlocal count
        count +=1
        return count

    return counter

counter_a = make_counter()
print(counter_a()) # 1
print(counter_a()) # 2
print(counter_a()) # 3

# nonlocal -> 바로 위 함수의 변수를 참조한다!
# 이렇게 함으로 순서는 항상 상승되는 보장할 수 있다.
# DB에서 게시물 번호
# 어디선가 DB 값을 만지지 못하게 함.
```

```python
# step2 : 파라미터를 설정합니다.
# 얻어가야할 포인트(데코레이터와는 관련은 없습니다.) : map은 __len__가 없어서 len()가 안됨.
# 포인트2: list 형변환은 부담이 있는 연사이니 주의를 해주세요.

def data_pre(function):
    def wrapper(iter_obj):
        return function(map(int, iter_obj))
    return wrapper

@data_pre
def mean(l):
    return sum(l) / len(l)

mean([1, 2, '3', 4, 5, '6']) # data_pre(mean)(iter_obj) # iter_obj에 [1, 2, '3', 4, 5, '6]

# 이 3.5는 실제 mean 반환값인가요? 실제는 wrapper의 반환값입니다.
```

# nonlocal

```python
# nonlocal은 파이썬의 키워드 중 하나로, 중첩 함수 내부에서 바깥 함수의 변수를 참조할 수 있게 합니다.
# nonlocal은 주로 클로저(closure)에서 변수의 값을 변경하고자 할 때 사용합니다.

# nonlocal
a = 10
def f():
    a = 100
    print(f'f a: {a}')
    def ff():
        a = 1000
        print(f'ff a: {a}')
        def fff():
            nonlocal a # global a로 변경해보세요.
            # global a
            a = 100
            print(f'fff a: {a}')
        fff()
        print(f'ff a: {a}')
    ff()
f()
print(f'global a: {a}')

# 내 밖에 있는 원소를 가져와서 수정할 때 활용
```

```python
import requests
# requests는 다양한 분야에서 사용이 됩니다. 테스트 용도로도 사용하고, 패킷 조작 등 통신에 관련된 것에 폭넓게 사용됩니다.

from bs4 import BeautifulSoup

paullab_url = 'https://paullab.co.kr/bookservice/'
response = requests.get(paullab_url)
response.encoding = 'utf-8'
html = response.text

soup = BeautifulSoup(html, 'html.parser')

bookservices = soup.select('.col-lg-6 > h2') # col-lg-6 클래스 안의 h2 태그 탐색
for no, book in enumerate(bookservices, 1):
    print(no, book.text)
```

```python
# 이 데이터를 DB에도 넣을 수 있고(import sqlite3)
# 파일로도 생성할 수 있는 데, 여러 형식으로 파일로 생성해보며 여러분과 얘기 나눠보도록 하겠습니다.
import requests
# requests는 다양한 분야에서 사용이 됩니다. 테스트 용도로도 사용하고, 패킷 조작 등 통신에 관련된 것에 폭넓게 사용됩니다.

from bs4 import BeautifulSoup

paullab_url = 'https://paullab.co.kr/bookservice/'
response = requests.get(paullab_url)
response.encoding = 'utf-8'
html = response.text

soup = BeautifulSoup(html, 'html.parser')
```

```python
s = ''
bookservices = soup.select('.col-lg-6 > h2') # col-lg-6 클래스 안의 h2 태그 탐색
bookcover = soup.select('.book_cover')

for no, book in enumerate(zip(bookservices, bookcover), 1):
    s += f'<h2>{no}. {book[0].text}</h2><img width="100px" src="https://paullab.co.kr/bookservice/{book[1]["src"]}"/>'

with open('test.html', 'w') as f:
    f.write(s)
```

```python
# csv로 뽑기
s = ''
bookservices = soup.select('.col-lg-6 > h2') # col-lg-6 클래스 안의 h2 태그 탐색
bookcover = soup.select('.book_cover')

for no, book in enumerate(zip(bookservices, bookcover), 1):
    s += f'{no}, {book[0].text}, "https://paullab.co.kr/bookservice/{book[1]["src"]}"\n'

with open('test.csv', 'w') as f:
    f.write(s)
```

```python
# json로 뽑기
# 따옴표가 붙지 않았는 데 \` 또는 \"로 각각 표기해주어야 합니다.
s = '['
bookservices = soup.select('.col-lg-6 > h2') # col-lg-6 클래스 안의 h2 태그 탐색
bookcover = soup.select('.book_cover')

for no, book in enumerate(zip(bookservices, bookcover), 1):
    s += f'{{number: {no}, bookname: {book[0].text}}},'

s = s[:-1] + ']'
s

with open('test.json', 'w') as f:
    f.write(s)
```

```python
print('hello')
try:
    s = 1/0
    print(s)
except:
    print('error가 발생되었습니다!')
print('world')
```

```python
try:
    s = 1/1 # 0에서 1로 값을 바꾸었습니다.
    print(s)
    l = [1, 2, 3]
    print(l[3])
except ZeroDivisionError:
    print('0으로 나누어졌습니다!')
except:
    print('오류가 났습니다!')
```
