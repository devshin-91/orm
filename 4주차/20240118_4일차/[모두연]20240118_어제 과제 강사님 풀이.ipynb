```python
import time

def time_decorator(func):
    def wrapper():
        start_time = time.time()
        result = func()
        end_time = time.time()
        print(f"함수가 {end_time - start_time} 초가 걸렸습니다.")
        return result
    return wrapper

@time_decorator
def f():
    s = 0
    for i in range(101):
        s += i
    return s

f()
```
