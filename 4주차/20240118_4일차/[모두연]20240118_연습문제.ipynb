# 문제 1: 이메일 주소 변경하기 (sub 사용)

```
- 문제: 주어진 문자열에서 모든 이메일 주소를 user@weniv.co.kr으로 변경하시오.

- 예시 입력: "저의 이메일 주소는 kim123@gmail.com입니다. 친구의 이메일 주소는 lee456@gmail.com입니다."
예시 출력: "저의 이메일 주소는 user@weniv.co.kr입니다. 친구의 이메일 주소는 user@weniv.co.kr입니다."

- 힌트: 이메일 주소를 찾기 위한 정규표현식은 \b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b를 사용할 수 있습니다.
```

# 문제 2: HTML 태그 제거하기 (sub 사용)

```
- 문제: 주어진 HTML 문자열에서 모든 HTML 태그를 제거하고 텍스트만 남기시오.

- 예시 입력: "<p>이것은 <b>예시</b> 문장입니다.</p>"
예시 출력: "이것은 예시 문장입니다."

- 힌트: HTML 태그를 찾기 위한 정규표현식은 <[^>]+>를 사용할 수 있습니다.
```

# 개인 답

````python
# 문제 1
'''
문제: 주어진 문자열에서 모든 이메일 주소를 user@weniv.co.kr으로 변경하시오.

예시 입력: "저의 이메일 주소는 kim123@gmail.com입니다. 친구의 이메일 주소는 lee456@gmail.com입니다." 예시 출력: "저의 이메일 주소는 user@weniv.co.kr입니다. 친구의 이메일 주소는 user@weniv.co.kr입니다."

힌트: 이메일 주소를 찾기 위한 정규표현식은 \b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+.[A-Z|a-z]{2,}\b를 사용할 수 있습니다.
'''

string1 = "저의 이메일 주소는 kim123@gmail.com입니다. 친구의 이메일 주소는 lee456@gmail.com입니다."
string2 = "저의 이메일 주소는 user@weniv.co.kr입니다. 친구의 이메일 주소는 user@weniv.co.kr입니다."

import re

def solution(my_string):
    return re.sub(r"@[A-Za-z0-9.-]+.[A-Z|a-z]{2,}", "@weniv.co.kr", my_string)

print(solution(string1))
print(solution(string2))
````

````python
# 문제 2

'''
- 문제: 주어진 HTML 문자열에서 모든 HTML 태그를 제거하고 텍스트만 남기시오.

- 예시 입력: "<p>이것은 <b>예시</b> 문장입니다.</p>"
예시 출력: "이것은 예시 문장입니다."

- 힌트: HTML 태그를 찾기 위한 정규표현식은 <[^>]+>를 사용할 수 있습니다.
'''


string = "<p>이것은 <b>예시</b> 문장입니다.</p>"

def solution(my_string):
    return re.sub(r"<[^>]+>", "", my_string)

print(solution(string))
````
