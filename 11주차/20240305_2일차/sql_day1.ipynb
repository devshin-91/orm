```python
# https://itholic.github.io/python-listdir-glob/

# 파일명 조회하여 파일명으로 db에 테이블 일괄 삽입

import os

path = "./jsons"
# os.listdir(path) # ['공급업체.json', '상품.json', '주문.json', '주문상세.json', '카테고리.json']
file_lists = os.listdir(path)
# file_lists[0].endswith(".json") # True

file_names = list()
for file in file_lists:
    if file.endswith(".json"):
        # print(file.split('.json')) # ['공급업체', '']
        # print(file.split('.json')[0]) # 공급업체
        file_name = file.split('.json')[0]
        file_names.append(file_name)
```

```python
file_names = [file.split('.json')[0] for file in file_lists if file.endswith(".json")]
```

```python
file_names
```

```python
# 커머스 db에 파일명에 해당하는 테이블을 일괄 삽입
conn = sqlite3.connect('커머스.db')
for name in file_names:
    with open(f'./jsons/{name}.json', encoding='UTF-8') as f:
        js = json.loads(f.read())
    f.close()
    df = pd.DataFrame(js)   
    df.to_sql(f'{name}', conn)
```
