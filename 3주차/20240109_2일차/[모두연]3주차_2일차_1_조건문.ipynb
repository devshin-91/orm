````python
# 리스트 컴프리헨션의 예
 [i for i in range(10) if i % 2 == 0] # [0, 2, 4, 6, 8]
[i if i % 2 == 0 else None for i in range(10)] # [0, None, 2, None, 4, None, 6, None, 8, None]
````

````python
# 조건문과 람다함수

# lambda arg: expression (조건문)
my_function = lambda x: 'Even' if x % 2 == 0 else 'Odd'

# map 응용
my_list = [1, 2, 3, 4, 5, 6]
print(list(map(my_function, my_list)))

# lambda의 예
# f(x) = x **2 -> 함수에 전달되는 인자: 표현식 -> x: x **2
제곱 = lambda x: x**2
res = 제곱(2)
# print(res)

num_list = [1, 2, 3, 4, 5]
print(list(map(제곱, num_list)))

# f(a, b) = a + b
덧셈 = lambda a,b : a + b
res = 덧셈(2, 3)
print(res)

length_list1 = [1, 2, 3, 4]
length_list2 = [5, 6, 7, 8]

res = list(map(덧셈, length_list1, length_list2))
print(res)
````

````python
# if, else 기본 구조

if 조건문: # 조건문 -> 참 -> if문 내 문장 수행
    # 수행할 문장 1
    pass

else: # if 안 조건문 -> 거짓 => else 문 실행
    # 수행할 문장 2
    pass
````

````python
# if, elif, else

if 조건문: # 조건문 -> 참 -> if문 수행
    pass
    # 수행할 문장 1
elif 추가조건: # if의 조건문 -> 거짓 -> elif 수행
    pass
    # 수행할 문장 2
else: # 위 모든 if문, elif문의 조건 -> 거짓 -> else 문 수행
    # 수행할 문장 3
    pass
````

````python
score = int(input("성적을 입력하세요 (0-100): ")) # step 1. 어떤 프로그램을 짤지

# 조건문의 범위는 겹치지 않게!
# 만일 조건문의 범위가 겹치면 데이터 처리가 잘못됨.

# step 2. 조건문 짜기
if score >= 90: # 점수가 90점 이상일 때
    print("A")
elif 90 > score >= 80: # 점수가 90점 미만 80점 이상
    print("B")
elif 80 > score >= 70: # 점수가 80점 미만 70점 이상
    print("C")
elif 70 > score >= 60: # 점수가 70점 미만 60점 이상
    print("D")
else: # 위 모든 조건이 거짓인 경우 (점수가 60점 미만)
    print("F")

# 테스트를 다양하게 각 케이스(조건) 테스트 수ㅎㅇ
# 음수, 소수점 등등
````

````python
# 조건문에 익숙하면 쓸 수 있는 조건 축약
score = int(input("성적을 입력하세요 (0-100): "))

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
````

````python
movie = {
    '영화': '레미제라블',
    '장르': '드라마',
    '서비스': '디즈니',
    '비용': 15000,
    '평점': 4.9
}

if movie['장르'] == '드라마': # 드라마
    if movie['서비스'] == '디즈니' or movie['서비스'] == '넷플릭스': # 디즈니 -> V -> 뒤 조건 안 봄
        if movie['평점'] >= 4.5: # 4.9 -> V
            if movie['비용'] <= 20000: # 15000 -> V
                print('영화시청') # 출력: 영화시청
````

````python
movie = {
    '영화': '레미제라블',
    '장르': '드라마',
    '서비스': '왓챠',
    '비용': 15000,
    '평점': 4.9
}

if movie['장르'] == 'SF': # 드라마 -> x
    if movie['서비스'] == '디즈니' or movie['서비스'] == '넷플릭스':
        if movie['평점'] >= 4.5 and movie['비용'] <= 20000:
            print('영화시청')
else: # 조건문 -> x -> else 수행
    print(movie['서비스'] + ' 구독하기!') # 출력: 왓챠 구독하기!
````
