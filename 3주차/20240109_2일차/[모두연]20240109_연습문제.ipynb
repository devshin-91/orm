# 개인 답

````python
# 문제 1
# 숫자를 하나 입력받아, 그 숫자가 5보다 큰지 확인하는 프로그램을 작성해보세요.

num = int(input('숫자를 입력해주세요: '))

def is_over5(num):
    if num < 5:
        res = '보다 작다'
    elif num > 5:
        res = '보다 크다'
    else:
        res = '와 같다'

    msg = f'5{res}'
    print(msg)

is_over5(num)

````

````python
# 문제 2
# "안녕하세요!"를 5번 출력하는 프로그램을 작성해보세요.

for _ in range(5):
    print("안녕하세요!")
````

````python
# 문제3
# 1부터 10까지의 숫자 중에서, 짝수만 출력하는 프로그램을 작성해보세요.

l = [ x  for x in range(1,11) if x % 2==0 ]
for even in l:
    print(even)
````

````python
# 문제4
# 1부터 20까지의 숫자 중, 3의 배수만 출력하는 프로그램을 작성하세요.

l = [ x  for x in range(1,21) if x % 3==0 ]
for mul_3 in l:
    print(mul_3)
````

````python
# 문제5
# 구구단 2단부터 9단까지 출력하는 프로그램을 작성하세요.

gugu = [ f'{i} x {j} = {i * j}' for i in range(2, 10) for j in range(1,10) ]
for i, val in enumerate(gugu):
    if i % 9 == 0:
        print(' ')
        print('==='*2+'*'+'==='*2)
        print(f'    {int(i/9) + 2}단')
        print('==='*2+'*'+'==='*2)
        print(' ')
    print(' '+val)
````
