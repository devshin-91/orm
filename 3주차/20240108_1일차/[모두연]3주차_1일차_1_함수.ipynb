# 2. 파라미터와 아규먼트

- 파라미터 : 함수를 정의할 때 사용되는 `변수`. 함수 호출 시, 들어온 값을 변수에 할당하는 역할 수행
- 아규먼트 : 함수를 호출할 떄 전달하는 `실제 값`

**함수를 정의할 때 설정된 파라미터의 수**와 **함수 호출 시 전달되는 아규먼트의 수**가 **일치**해야함

````python
# 파라미터의 기본 사용
def f(a, b, c):
    print(a, b, c)

f(100,200,300)
# 키워드 인자(keyword arguments ; a, b, c, ....)를 사용하여 파라미터에 값을 할당
# f(a=100,300,200) # 에러 SyntaxError: positional argument follows keyword argument
f(a=100, b=200, c=300) # 출력 : 100 200 300
f(c=300, b=200, a=100) # 출력 : 100 200 300
````

````python
# f(a=100,300,200)
# SyntaxError: SyntaxError: positional argument follows keyword argument
````

````python
# 기본값을 가지는 파라미터

# 잘못된 예
# def f(a=10, b=20, c): # 기본값이 없는 파라미터가 기본값이 있는 파라미터 뒤에 올 수 없음
#     print(a, b, c) # error  name 'c' is not defined

# 올바른 예
def f(a=10, b=10, c=30):
    print(a, b, c)

f() # 출력 : 10 10 30 -> 이미 함수에서 키워드 인자에 기본값들이 설정되어있어서, 아규먼트가 생략되어도 함수가 기본값들을 사용.
f(100, 10)
f(a=100, b=200, c=300)
f(c=300, a=100, b=200)
````

````python
# 일부 파라미터만 기본 값을 가지는 경우
def f(a, b=20, c=30): #
    return a, b, c

# print(f()) # error
print(f(100, 10))
print(f(a=100, b=200, c=300))
print(f(c=300, b=100, a=200))
# print(f(c=300)) # error
````

````python
# 권장하지 않는 방법
def append_to_list(value, lst=[]):
    lst.append(value)
    return lst # 예상치 못한 결과가 발생할 수 있다.
````

````python
append_to_list(1) # 출력: [1]
````

````python
append_to_list(2) # 출력 : [2]? 근데 [1,2 ]. 앞에서 출력이 lst = [1]로 lst가 업데이트 되어서.
````

````python
# 기본값이 변경 가능한 객체일 때는 유의하자.
# 권장 방법
def append_to_list(value, lst=None):
    if lst is None:
        lst = []
    lst.append(value)
    return lst
````

````python
append_to_list(1) # 출력: [1]
````

````python
append_to_list(2) # 출력 : [2]
````
