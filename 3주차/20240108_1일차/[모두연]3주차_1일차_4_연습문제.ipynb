```python
a = 'pithon'

def 함수1():
    def 함수2():
        print('love')

    print('I') # 1번째 출력 - 함수1() 내 출력이어서 제일 먼저 출력
    함수2() # 2번째 - 함수2()는 함수1() 내 지역함수여서 2번째로 출력
    return "python" # 3번째 - 마지막으로 python이 반환되어서 3번쨰 출력

a = 함수1()
print(a)

# I
# love
# python # 반환값으로 출력됨.
```

```python
# 함수를 이용하여 계산기 프로그램을 만들어주세요.
```

```python
def plus(num1, num2):
    # 덧셈
    return num1 + num2

def minus(num1, num2):
    # 뺄셈
    return num1 - num2

def multiply(num1, num2):
    # 곱셈
    return num1 * num2

def divide(num1, num2):
    # 나눗셈
    if num2 == 0: # 0으로 나눌 경우 divisionerror가 발생
        print("Don't use 0")
    else:
        return num1 / num2

print(f'plus : {plus(10, 5)}')
print(f'minus : {minus(10, 5)}')
print(f'multiply : {multiply(10, 5)}')
print(f'divide : {divide(10, 5)}')
```
