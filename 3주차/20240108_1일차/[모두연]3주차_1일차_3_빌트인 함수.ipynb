Built in function

## 1. all, any

```python
# 1. all,  any


# all은 모두가 True일 때 True. 마치 and와 비슷
all([True, True, True]) # 출력: True
all([True, True, False]) # 출력: False
all([1, 1, 1, 0]) # 출력: False
all([1, 0, 1, 1]) # 출력: False
all([1, 100, 10, -1]) # 출력: True
```

```python
# any는 하나라도 True면 True. 마치 or과 비슷
any([True, False, False]) # 출력: True
```

```python
any([False, False, False]) # 출력: False
```

```python
# Map 함수
```

```python
# 기본 사용법 : map(함수, 리스트 혹은 튜플)
```

```python
# 첫번째 방법
def 제곱(x):
    return x ** 2

list(map(제곱, [1, 2, 3, 4])) # 출력: [1, 4, 9, 16]
```

```python
# 왜 map을 써야하나?
# 리스트의 경우 리스트 자체에 리스트의 원소 각각을 제어하는 메서드가 없어서
# [1, 2, 3, 4] ** 2 # error : unsupported operand type(s) for ** or pow(): 'list' and 'int'
# [1, 2, 3, 4] * 2 # 출력 : [1, 2, 3, 4, 1, 2, 3, 4]
```

```python
# 두번째 방법
list(map(lambda x: x **2, [1, 2, 3, 4]))
```

```python
# Filter 함수
```

```python
# 기본 사용 : filter(조건 함수, 리스트 또는 튜플)
```

```python
list(filter(lambda x : x > 50, range(100)))
```

```python
# zip 함수
```

```python
# 기본 사용법 : zip(순회 가능 객체1, 순회 가능 객체2...)
# 각 자료형의 인덱스 순번끼리 묶음
list(zip('ab', '12')) # 출력: [('ab'[0], '12'[0]), ('ab'[1], '12'[1])]-> [('a', '1'), ('b', '2')]
```

```python
list(zip('abc', '1234567', [10, 20, 30, 40]))
# 출력 : [('a', '1', 10), ('b', '2', 20), ('c', '3', 30)]
```

```python
# 용도 : 알고리즘, 좌표 접근에 사용
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
list(zip(x,y))
```

```python
# print, repr
```

```python
class Charactor:
    def __repr__(self): # 해당 문자열을 가져올 때 씀
        return 'hello'
    def __str__(self):
        return 'world'

hero = Charactor()
print(hero) # 출력 : 'world'
hero # 출력 : 'hello'
```

```python
# enumerate - 순회가능한 객체를 이용 -> 반복문에서 주로 볼 수 있음.
# 용도
# index를 자동으로 생성할 때 자주 사용.
# 리스트는 키가 없지만 필요에 따라 enumerate로 임시 키값을 생성
# 이때, 키값을 바라보는 서로 다른 데이터를 묶을 때 사용
```

```python
values = ['A', 'B', 'C', 'D']

print(list(enumerate(values)))
print(list(enumerate(values, 1))) # 인덱스 1부터 시작
print(list(enumerate(values, 100))) # 인덱스 100부터 시작
```

```python
values = ['김진환', '이호준', '라이켓', '웅이']
print(list(enumerate(values)))
```
