# 1. 지역변수

```python
# 지역 변수 사용 예시
def f():
    a = 1
    print(a)  # 출력: 1

f()
print(a)  # 에러:  a는 f() 함수 내에서만 정의되어서 밖에서는 print가 a를 찾지 못함
```

```python
# 그럼 a를 바깥에서 사용하려면?
'''
1. a를 반환을 해준다.
'''
```

```python
# locals() : 지역변수 확인하고자할 때 쓰는 함수

def f():
    a = 10
    b = 100
    print(locals()) # 출력: {'a': 1, 'b': 100}

f()
```

```python
def f():
    a = 1
    b = 'hello'
    def ff(): # 함수 안에서 함수를 사용가능하나 그 함수 자체도 지역 변수가 될 수 있다.
        pass
    print(locals()) # 출력: {'a': 1, 'b': 'hello', 'ff': <function f.<locals>.ff at 0x7fbde95ebb50>}

f()
```

# 전역변수

```python
# 전역변수 사용 권장 x
# 이유
# 1. global 키워드 사용한 전역 변수의 변경이 권장 되지 않음
# 2. 기본값이 변경될 수 있음.
```

```python
# global 함수 => 전역변수 지정
a = 100

def f():
    global a
    a = a + 1
    return a

print(f()) # 101
print(a) # 함수 내에서 전역변수 a가 수정됨.
```

```python
# 전역변수와 지역변수 구분
a = 100

def f():
    a = 1000
    print(a)

f() # f() 함수 내에 지역변수 a가 정의되어 있어 지역변수 a를 출력함.
```
