#2. 시퀀스 자료형의 패킹과 언패킹

```python
# 패킹 : 그룹화
# 언패킹 : 분해
```

```python
# 2.1 튜플 패킹

# 튜플 패킹
packed_tuple = 1, 2, 3, 4
print(packed_tuple) # 출력: (1, 2, 3,4)
type(packed_tuple)
```

```python
tuple_example = 1
print(tuple_example)
type(tuple_example)
```

```python
# 2.2 언패킹
```

```python
# 튜플 언패킹

# a, b, c, d = packed_tuple
# a, c, b, d = packed_tuple
a, b, c, d = packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
# 다양한 패킹/언패킹

# range : 0 부터 n-1까지 숫자를 생성하는 함수

# range(3)
# a = range(3) # 0, 1, 2
# print(a)

a, b, c = range(3) # 0, 1, 2
print(range(3))
print(a,b,c) # 언패킹을하면 순차적으로 숫자가 나옴
```

```python
a, b, c, d, e = 'hello'
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
a, b = [10, 20]
print(a)
print(b)
```

```python
# 언패킹은 시퀀스 자료형이라면 전부 활용 가능
```

```python
total_sum = 0

for i in range(5):
    total_sum += i
    print(total_sum)

print(total_sum)
```

```python
# 별표를 이용한 언패킹
first, *middle, last = [1,2,3,4,5]
print(first)
print(middle) # 출력: [2, 3, 4]
print(last)
```

```python
# 별표를 이용한 언패킹
first, *middle = [1,2,3,4,5]
*middle, first = [1,2,3,4,5]
print(first)
print(middle) # 출력: [1, 2, 3, 4]
```
