```python
# 중복제거
set('123123123123')
```

```python
# 문자형태의 숫자를 중복제거하고 그걸 다 더하고 싶어
# * 접근 방법
# set : 문자열의 중복값 제거 -> map으로 원소들을 int로 형변환

# 1. map을 이용해서 중복 제거가 된 원소값들을 일괄적으로 int로 변환
map(int, set('123123123123')) # 다만 출력은 안 되고 메모리 상 올라감
# 2. sum으로 원소들을 다 합침
sum(map(int, set('123123123123'))) # 원소들을 다 합쳐서 출력
```

```python
# sum(순회가능한객체, 초깃값)
sum([10, 20, 30], 100)
```

```python
# 모든 원소의 합은 어떻게 구할까?
sum([[10, 20], [30, 40], [50, 60]], []) # 평탄화 작업)
# [] + [10, 20] == [10, 20]
# [10, 20] + [30, 40] == [10, 20, 30, 40]
# => 1차원으로 평탄화 시켜서 리스트 원소를 모두 더함.
```

```python
# 평탄화된 리스트를 다시 sum으로 합치면 모든 원소의 합을 구할 수 있다.
```
