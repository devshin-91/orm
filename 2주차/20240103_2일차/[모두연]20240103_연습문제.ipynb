# 1번 문제
다음 변수들이 가리키고 있는 값에 화살표를 그려주세요.

(z의 값은 무엇일까요?)

```python
# 문제1
x = 100
y = 100
z = x
```

```python
# 답
'''
z -> x -> 100
y -> 100
'''
'z의 값은 100'

# 메모리 상 위치
print(id(x), id(y), id(z))
```

```python
# 문제2
x = 258
y = 258
z = x
```

```python
# 답
'''
z - > x -> 258
y -> 258
'''
'z의 값은 258'

# 메모리 상 위치
print(id(x), id(y), id(z))
```

# 6번 문제

'abcde', 숫자, 대쉬(-), 언더스코어(_)를 모두 제거해주세요. 문자열에 translate를 사용해주세요.

```python
s = 'hello_world123'
# 출력값: 'hlloworl'
```

```python
# 답
s = 'hello_world123'

table = s.maketrans('','','abcde123456789-_')
res = s.translate(table)
print(res)
```

# 7번 문제

확장자를 출력해주는 코드를 작성해주세요. 확장자는 png, jpeg, mp4 등입니다.

```python
s = 'licat.jpeg'
# 출력값: 'jpeg'

s = 'li.cat.png'
# 출력값: 'png'
```

```python
# 답

'''
# 접근 방법
# .이 몇 개 가 나오든 .을 구분자로 split했을 때 마지막에 나오는 원소가 확장자
'''
# s = 'licat.jpeg'
# s = 'li.cat.png'

ext_s = s.split(sep='.')[-1]
print(ext_s)
```
