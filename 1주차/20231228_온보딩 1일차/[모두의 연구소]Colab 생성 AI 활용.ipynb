```python
# prompt : 빙고 게임을 만드려고하는데 5*5형식으로 만드는 코드를 작성해줘

import random

# Create a 5x5 grid of numbers
grid = [[0 for i in range(5)] for j in range(5)]

# Fill the grid with numbers from 1 to 25
for i in range(5):
  for j in range(5):
    grid[i][j] = i * 5 + j + 1

# Choose a random number to be the winning number
winning_number = random.randint(1, 25)

# Play the game
while True:
  # Get a guess from the user
  guess = int(input("Enter your guess: "))

  # Check if the guess is correct
  if guess == winning_number:
    print("You win!")
    break
  else:
    print("Try again.")

# Print the winning number
print("The winning number was", winning_number)
```
